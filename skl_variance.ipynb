```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define ranges for x and y
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Compute z
Z = np.log(X / (0.8 * X + 0.2 * Y)) + 1 - 0.8 * (X / (0.8 * X + 0.2 * Y))

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D plot of z = log(x / (0.8x + 0.2y)) + 1 - 0.8 * (x / (0.8x + 0.2y))')

plt.tight_layout()
plt.show()


# Create a contour plot of the same function
fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
cbar = plt.colorbar(contour)
cbar.set_label('z value')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Contour plot of z = log(x / (0.8x + 0.2y)) + 1 - 0.8 * (x / (0.8x + 0.2y))')

plt.tight_layout()
plt.show()
```

```python
# Define ranges for x and y
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

alpha = 0.5

# Compute z
Z = np.log(X / (alpha * X + (1-alpha) * Y)) + 1 - alpha * (X / (alpha * X + (1-alpha) * Y))

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D plot of z = log(x / (0.8x + 0.2y)) + 1 - 0.8 * (x / (0.8x + 0.2y))')

plt.tight_layout()
plt.show()

# Create a contour plot of the same function
fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
cbar = plt.colorbar(contour)
cbar.set_label('z value')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Contour plot of z = log(x / (0.5x + 0.5y)) + 1 - 0.5 * (x / (0.5x + 0.5y))')

plt.tight_layout()
plt.show()
```

```python
# Define ranges for x and y
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Compute z
Z = np.log(X / Y)

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D plot of z = log(x / y)')

plt.tight_layout()
plt.show()


# Create a contour plot of the same function
fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
cbar = plt.colorbar(contour)
cbar.set_label('z value')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Contour plot of z = log(x / y)')

plt.tight_layout()
plt.show()
```
